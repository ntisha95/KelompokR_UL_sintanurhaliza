```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import seaborn as sns
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(10, 5))
linked = linkage(X_scaled, method='ward')
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Dendrogram untuk Iris Dataset')
plt.xlabel('Sampel')
plt.ylabel('Jarak')
plt.show()
```

```python
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster = agg.fit_predict(X_scaled)
X['Cluster'] = cluster
```

```python
sns.scatterplot(data=X, x=iris.feature_names[0], y=iris.feature_names[1], hue='Cluster', palette='Set1')
plt.title('Agglomerative Clustering pada Iris Dataset')
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
print("Adjusted Rand Index:", adjusted_rand_score(y, cluster))
```
